Let's calculate the norm.

```python
nm_norm = sqrt(not_moving['x']**2+not_moving['y']**2 +not_moving['z']**2)
mo_norm = sqrt(moving['x']**2+moving['y']**2 +moving['z']**2)
```

```python
nm_norm.plot()
```

```python
mo_norm.plot()
```

Let's calculate some features

```python
from pandas.stats.moments import rolling_mean, rolling_var, rolling_std

mo_mean = rolling_mean(mo_norm, 50)
nm_mean = rolling_mean(nm_norm, 50)

mo_var = rolling_var(mo_norm, 50)
nm_var = rolling_var(nm_norm, 50)

mo_std = rolling_std(mo_norm, 50)
nm_std = rolling_std(nm_norm, 50)
```

```python
mo_mean.plot()
figure()
nm_mean.plot()
```

```python
import pandas
mo_features = []
nm_features = []

for i in range(0,len(mo_mean)-1):
    if not(pandas.isnull(mo_mean[i])):
        mo_features.append([mo_mean[i],mo_var[i],mo_std[i]])

for i in range(0,len(nm_mean)-1):
    if not(pandas.isnull(nm_mean[i])):
        nm_features.append([nm_mean[i],nm_var[i],nm_std[i]])

```
